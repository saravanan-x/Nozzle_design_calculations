```python
import math

def get_atmosphere_data(h):
    """
    Calculates atmospheric properties based on altitude (h) in meters.
    Returns: temperature (C), pressure (kPa), density (kg/m³)
    """

    if h < 11000:
        # Troposphere
        temp = 15.04 - 0.00649 * h
        pressure = 101.29 * ((temp + 273.1) / 288.08)**5.256

    elif 11000 <= h <= 25000:
        # Lower Stratosphere
        temp = -56.46
        pressure = 22.65 * math.exp(1.73 - 0.000157 * h)

    else:
        # Upper Stratosphere
        temp = -131.21 + 0.00299 * h
        pressure = 2.488 * ((temp + 273.1) / 216.6)**-11.388

    density = pressure / (0.2869 * (temp + 273.1))

    return temp, pressure, density


# Example
Target_height = 25 #KM
altitude = Target_height * 1000   # Requuired KM of rocket

t, p, rho = get_atmosphere_data(altitude)

temp_k = t + 273.15
pressure_pa = p * 1000
pressure_mpa = pressure_pa / 1e6

print(f"Altitude: {altitude} m")
print(f"Temperature: {t:.2f} °C")
print(f"Temperature: {temp_k:.2f} K")
print(f"Pressure: {p:.2f} kPa")
print(f"Pressure: {pressure_pa:.2f} Pa")
print(f"Pressure: {pressure_mpa:.6f} MPa")
print(f"Density: {rho:.4f} kg/m³")
```

```python
import math

# INPUT PARAMETERS

F = 5000                 # Thrust (N)
P1 = 2.068e6             # Chamber pressure (Pa)
P2 = 0.002549e6          # Exit pressure (Pa) (ambient at 25 km)
T1 = 2800                # Chamber temperature (K)

k = 1.30                 # Specific heat ratio
R = 355.4                # Gas constant (J/kg-K)

# PRESSURE RATIO
pressure_ratio = P2 / P1
print("Pressure ratio P2/P1 =", pressure_ratio)


# CRITICAL PRESSURE (THROAT)
Pt = ((2/(k+1))**(k/(k-1))) * P1
print("Critical pressure Pt =", Pt/1e6, "MPa")


# THROAT VELOCITY
vt = math.sqrt((2*k/(k+1)) * R * T1)
print("Throat velocity vt =", vt, "m/s")


# EXIT VELOCITY
ve = math.sqrt((2*k/(k-1)) * R * T1 * (1 - (P2/P1)**((k-1)/k)))
print("Exit velocity ve =", ve, "m/s")


# MASS FLOW RATE
m_dot = F / ve
print("Mass flow rate =", m_dot, "kg/s")

# SPECIFIC VOLUME AT CHAMBER
V1 = (R * T1) / P1
print("Specific volume V1 =", V1, "m^3/kg")


# THROAT SPECIFIC VOLUME
Vt = V1 * ((k+1)/2)**(1/(k-1))
print("Throat specific volume Vt =", Vt, "m^3/kg")


# EXIT SPECIFIC VOLUME
V2 = V1 * (P1/P2)**(1/k)
print("Exit specific volume V2 =", V2, "m^3/kg")


# THROAT AREA
At = (m_dot * Vt) / vt
print("Throat area At =", At, "m^2")
print("Throat area =", At*10000, "cm^2")


# EXIT AREA
Ae = (m_dot * V2) / ve
print("Exit area Ae =", Ae, "m^2")
print("Exit area =", Ae*10000, "cm^2")

# EXPANSION RATIO
epsilon = Ae / At
print("Expansion ratio Ae/At =", epsilon)

# EXIT TEMPERATURE
T2 = T1 * (P2/P1)**((k-1)/k)
print("Exit temperature =", T2, "K")
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# -----------------------------
# Parameters
# -----------------------------
f = 1200          # Thrust (N)
pc = 2.468e6      # Chamber pressure (Pa)
pe = 0.1e6        # Exit pressure (Pa)
tc = 2800         # Chamber temperature (K)
y = 1.3           # Specific heat ratio
r = 355.4         # Gas constant (J/kgK)
pi = np.pi
ε = 5.23          # Area ratio (initial guess, not used directly)

# angles
divergent_angle = 15
convergent_a = 30

da = np.radians(divergent_angle)
ca = np.radians(convergent_a)

# characteristic length
length = 0.7

# -----------------------------
# STEP 1: Exit velocity
# -----------------------------
ve = np.sqrt((2*y/(y-1))*r*tc*(1-(pe/pc)**((y-1)/y)))
print(f"Exit velocity: {ve:.2f} m/s")

# -----------------------------
# STEP 2: Mass flow rate
# -----------------------------
m = f / ve
print(f"Mass flow rate: {m:.4f} kg/s")

# -----------------------------
# STEP 3: Throat area
# -----------------------------
at = (m/pc) * np.sqrt((r*tc)/y * ((y+1)/2)**((y+1)/(2*(y-1))))
print(f"Area of throat: {at:.6f} m^2")

# throat diameter
dt = np.sqrt((4*at)/pi)

# -----------------------------
# STEP 4: Exit Mach
# -----------------------------
me = np.sqrt((2/(y-1))*((pc/pe)**((y-1)/y)-1))
print(f"Exit Mach number: {me:.2f}")

# -----------------------------
# STEP 5: Area ratio (A/A*)
# -----------------------------
Ar = ((y+1)/2)**(-(y+1)/(2*(y-1))) * (
    (1 + ((y-1)/2)*me**2)**((y+1)/(2*(y-1))) / me
)
print(f"ratio of area: {Ar:.4f}")

# -----------------------------
# STEP 6: Exit area
# -----------------------------
Aa = at * Ar
print(f"Exit Area: {Aa:.6f} m^2")

dee = np.sqrt((4*Aa)/pi)
print(f"Exit Dia: {dee*1000:.2f} mm")

# -----------------------------
# (Optional check - not used further)
# -----------------------------
cp = pc
ct = tc

mg = ((at * cp) / np.sqrt(ct)) * np.sqrt(y / r) * (
    me * (1 + ((y - 1) / 2) * me**2)**(-(y + 1) / (2 * (y - 1)))
)
print(f"General mass flow rate (check): {mg:.4f} kg/s")

mc = ((at * cp) / np.sqrt(ct)) * np.sqrt(y / r) * (
    ((y + 1) / 2)**(-(y + 1) / (2 * (y - 1)))
)
print(f"Choked mass flow rate: {mc:.4f} kg/s")

# -----------------------------
# FINAL MASS FLOW (use thrust-based)
# -----------------------------
# DO NOT overwrite m
# m = mg  ❌ WRONG

# -----------------------------
# Exit area final
# -----------------------------
ae = Aa
print(f"Area of Exit: {ae:.6f} m^2")

de = np.sqrt((4*ae)/pi)

# -----------------------------
# Geometry
# -----------------------------
cd = 3 * dt
ac = (pi*(cd**2))/4

lc = (length * at) / ac

convergent_length = (cd - dt) / (2*np.tan(ca))
divergent_length = (de - dt) / (2*np.tan(da))

total_length = lc + convergent_length + divergent_length

# -----------------------------
# Convert to mm
# -----------------------------
dt_mm = dt * 1000
de_mm = de * 1000
cd_mm = cd * 1000
lc_mm = lc * 1000
conv_mm = convergent_length * 1000
div_mm = divergent_length * 1000
total_mm = total_length * 1000

# -----------------------------
# Print Geometry
# -----------------------------
print(f"\n--- GEOMETRY ---")
print(f"Chamber diameter: {cd_mm:.2f} mm")
print(f"Chamber length: {lc_mm:.2f} mm")
print(f"Convergent length: {conv_mm:.2f} mm")
print(f"Throat diameter: {dt_mm:.2f} mm")
print(f"Divergent length: {div_mm:.2f} mm")
print(f"Exit diameter: {de_mm:.2f} mm")
print(f"Total length: {total_mm:.2f} mm")

# -----------------------------
# FLOW PROPERTIES
# -----------------------------
ρ0 = pc / (r * tc)
print(f"\nStagnation Density: {ρ0:.2f} kg/m^3")

vt = np.sqrt((2*y*r*tc)/(y+1))
print(f"Throat Velocity: {vt:.2f} m/s")

ρt = ρ0 * (2/(y+1))**(1/(y-1))
print(f"Throat Density: {ρt:.2f} kg/m^3")

# Exit pressure (correct)
ep = pc / (1 + ((y-1)/2)*me**2)**(y/(y-1))
print(f"Exit Pressure: {ep:.2f} Pa")

# Exit temperature
et = tc / (1 + ((y-1)/2)*me**2)
print(f"Exit Temperature: {et:.2f} K")

# Exit velocity (Mach-based check)
v_e = me * np.sqrt(y * r * et)
print(f"Exit velocity (check): {v_e:.2f} m/s")

# -----------------------------
# Thrust calculation
# -----------------------------
pa = pe   # assume ideally expanded

F = m * ve + (ep - pa) * ae
print(f"\nFinal Thrust: {F:.2f} N")
```
