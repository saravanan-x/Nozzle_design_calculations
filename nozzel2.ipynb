```python
import math

def get_atmosphere_data(h):
    """
    Calculates atmospheric properties based on altitude (h) in meters.
    Returns: temperature (C), pressure (kPa), density (kg/m³)
    """

    if h < 11000:
        # Troposphere
        temp = 15.04 - 0.00649 * h
        pressure = 101.29 * ((temp + 273.1) / 288.08)**5.256

    elif 11000 <= h <= 25000:
        # Lower Stratosphere
        temp = -56.46
        pressure = 22.65 * math.exp(1.73 - 0.000157 * h)

    else:
        # Upper Stratosphere
        temp = -131.21 + 0.00299 * h
        pressure = 2.488 * ((temp + 273.1) / 216.6)**-11.388

    density = pressure / (0.2869 * (temp + 273.1))

    return temp, pressure, density


# Example
Target_height = 25 #KM
altitude = Target_height * 1000   # Requuired KM of rocket

t, p, rho = get_atmosphere_data(altitude)

temp_k = t + 273.15
pressure_pa = p * 1000
pressure_mpa = pressure_pa / 1e6

print(f"Altitude: {altitude} m")
print(f"Temperature: {t:.2f} °C")
print(f"Temperature: {temp_k:.2f} K")
print(f"Pressure: {p:.2f} kPa")
print(f"Pressure: {pressure_pa:.2f} Pa")
print(f"Pressure: {pressure_mpa:.6f} MPa")
print(f"Density: {rho:.4f} kg/m³")
```

```python
import math

# INPUT PARAMETERS

F = 5000                 # Thrust (N)
P1 = 2.068e6             # Chamber pressure (Pa)
P2 = 0.002549e6          # Exit pressure (Pa) (ambient at 25 km)
T1 = 2800                # Chamber temperature (K)

k = 1.30                 # Specific heat ratio
R = 355.4                # Gas constant (J/kg-K)

# PRESSURE RATIO
pressure_ratio = P2 / P1
print("Pressure ratio P2/P1 =", pressure_ratio)


# CRITICAL PRESSURE (THROAT)
Pt = ((2/(k+1))**(k/(k-1))) * P1
print("Critical pressure Pt =", Pt/1e6, "MPa")


# THROAT VELOCITY
vt = math.sqrt((2*k/(k+1)) * R * T1)
print("Throat velocity vt =", vt, "m/s")


# EXIT VELOCITY
ve = math.sqrt((2*k/(k-1)) * R * T1 * (1 - (P2/P1)**((k-1)/k)))
print("Exit velocity ve =", ve, "m/s")


# MASS FLOW RATE
m_dot = F / ve
print("Mass flow rate =", m_dot, "kg/s")

# SPECIFIC VOLUME AT CHAMBER
V1 = (R * T1) / P1
print("Specific volume V1 =", V1, "m^3/kg")


# THROAT SPECIFIC VOLUME
Vt = V1 * ((k+1)/2)**(1/(k-1))
print("Throat specific volume Vt =", Vt, "m^3/kg")


# EXIT SPECIFIC VOLUME
V2 = V1 * (P1/P2)**(1/k)
print("Exit specific volume V2 =", V2, "m^3/kg")


# THROAT AREA
At = (m_dot * Vt) / vt
print("Throat area At =", At, "m^2")
print("Throat area =", At*10000, "cm^2")


# EXIT AREA
Ae = (m_dot * V2) / ve
print("Exit area Ae =", Ae, "m^2")
print("Exit area =", Ae*10000, "cm^2")

# EXPANSION RATIO
epsilon = Ae / At
print("Expansion ratio Ae/At =", epsilon)

# EXIT TEMPERATURE
T2 = T1 * (P2/P1)**((k-1)/k)
print("Exit temperature =", T2, "K")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# INPUT PARAMETERS
# =========================
pc = 3.468e6       # Pa
tc = 3000          # K
y = 1.22
r = 355.4

F_target = 5000    # N
pe = 0.02e6        # Pa
pa = 101325        # ambient

# angles
theta_n = np.radians(30)   # bell start
theta_e = np.radians(7)    # exit
theta_c = np.radians(30)   # convergent

# characteristic length
L_star = 0.5

# =========================
# INITIAL GUESS
# =========================
at = 1e-4   # initial throat area guess

# =========================
# ITERATE TO MATCH THRUST
# =========================
for i in range(10):
    # mass flow (choked)
    m = (pc * at / np.sqrt(tc)) * np.sqrt(y / r) * ((2/(y+1))**((y+1)/(2*(y-1))))
    
    # exit Mach
    me = np.sqrt((2/(y-1))*(((pc/pe)**((y-1)/y))-1))
    
    # exit area
    ae = (at/me) * (((1+((y-1)/2)*(me**2)) / ((y+1)/2))**((y+1)/(2*(y-1))))
    
    # exit velocity
    ve = np.sqrt((2*y/(y-1))*r*tc*(1-(pe/pc)**((y-1)/y)))
    
    # thrust
    F = m * ve + (pe - pa) * ae
    
    # scale throat area
    at = at * (F_target / F)

# =========================
# FINAL GEOMETRY
# =========================
dt = np.sqrt((4*at)/np.pi)
de = np.sqrt((4*ae)/np.pi)

rt = dt / 2
re = de / 2

# =========================
# COMBUSTION CHAMBER
# =========================
cd = 3 * dt
rc = cd / 2

ac = np.pi * rc**2
lc = (L_star * at) / ac

# =========================
# CHAMBER
# =========================
x_chamber = np.linspace(-lc, 0, 50)
r_chamber = np.full_like(x_chamber, rc)

# =========================
# CONVERGENT (LINEAR - STABLE)
# =========================
x_conv = np.linspace(0, 20e-3, 50)   # 20 mm length (adjustable)
r_conv = np.linspace(rc, rt, 50)

# shift to connect chamber
x_conv = x_conv + x_chamber[-1]

# =========================
# THROAT ARC (ADJUSTED)
# =========================
Rc = 1.5 * rt
theta_n = np.radians(30)

theta = np.linspace(0, theta_n, 100)

x_throat = Rc * np.sin(theta)
r_throat = rt + Rc * (1 - np.cos(theta))

# shift to match convergent end
x_throat = x_throat + x_conv[-1]
r_throat = r_throat + (r_conv[-1] - r_throat[0])

# =========================
# BELL NOZZLE (RAO)
# =========================
L_cone = (de - dt)/(2*np.tan(np.radians(15)))
L = 0.6 * L_cone

x_par = np.linspace(0, L, 100)

x_n = x_throat[-1]
r_n = r_throat[-1]

r_par = r_n + x_par*np.tan(theta_n) + (x_par**2/(2*L))*(np.tan(theta_e) - np.tan(theta_n))
x_par = x_par + x_n

# =========================
# COMBINE
# =========================
x_total = np.concatenate([x_chamber, x_conv, x_throat, x_par])
r_total = np.concatenate([r_chamber, r_conv, r_throat, r_par])

# =========================
# PLOT
# =========================
plt.figure(figsize=(8,4))
plt.plot(x_total*1000, r_total*1000, linewidth=2)

# mirror bottom
plt.plot(x_total*1000, -r_total*1000, linewidth=2)

plt.xlabel("Length (mm)")
plt.ylabel("Radius (mm)")
plt.title("Complete Rocket Engine Geometry (Chamber + Bell Nozzle)")
plt.axis('equal')
plt.grid()
plt.show()

# =========================
# OUTPUT
# =========================
print(f"Thrust: {F:.2f} N")
print(f"Mass flow: {m:.4f} kg/s")
print(f"Throat diameter: {dt*1000:.2f} mm")
print(f"Exit diameter: {de*1000:.2f} mm")
print(f"Chamber length: {lc*1000:.2f} mm")
```
